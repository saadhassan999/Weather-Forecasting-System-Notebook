# 1. Reading Weather Data 

```python
import pandas as pd
```

```python
rwp_data = pd.read_csv("Rawalpindi.csv")
```

```python
rwp_data.head()
```

```python
rwp_data.info()
```

```python
rwp_data.describe()
```

# 2. Data Cleaning 
       (Convert date time format [2020-11-12] to [20201112])

```python
rwp_data['date_time'] = rwp_data['date_time'].astype(str).str.replace(r'[^\d]', '', regex=True)
rwp_data.head()
```

# 3. Finding Correlation Between Features 

```python
corr_matrix = rwp_data.corr()
corr_matrix['tempC'].sort_values(ascending=False)
```

# 4. Defining Label and Features

```python
features = rwp_data[['maxtempC', 'WindChillC', 'FeelsLikeC', 'HeatIndexC', 'mintempC', 'uvIndex', 'pressure']]
label = rwp_data['tempC']
```

# 5. Splitting data into training and test set

```python
from sklearn.model_selection import train_test_split
features_train, features_test, label_train, label_test = train_test_split(features, label, test_size=0.2, random_state=42)
# print(f"rows in features training: {len(features_train)}\nrows in features test: {len(features_test)}\nrows in label training: {len(label_train)}\nrows in label test: {len(label_test)}")
```

# 6. Training algorithm (Random Forest)

```python
from sklearn.ensemble import RandomForestRegressor
train_model = RandomForestRegressor(n_estimators=500, random_state=0)
train_model.fit(features_train, label_train) 
```

# 7. Predicting Results

```python
predict_model = train_model.predict(features_train)
```

```python
# predict_model.tolist()
```

```python
# label_train.tolist()
```

# 8. Evaluating Model

```python
from sklearn import metrics
import numpy as np
print('Mean Absolute Error:', metrics.mean_absolute_error(label_train, predict_model))
print('Mean Squared Error:', metrics.mean_squared_error(label_train, predict_model))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(label_train, predict_model)))
```

# 9. Cross Validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(train_model, features_train, label_train, scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
def print_scores(scores):
    print("Mean: ", scores.mean())
    print("Standard deviation: ", scores.std())
print_scores(rmse_scores)
```

# 10. Saving Model

```python
from joblib import dump, load
dump(train_model, 'RFRegressor.joblib')
```

# 11. Testing Model

```python
final_predictions = train_model.predict(features_test)
print(final_predictions, list(label_test))
print('Mean Absolute Error:', metrics.mean_absolute_error(label_test, final_predictions))
print('Mean Squared Error:', metrics.mean_squared_error(label_test, final_predictions))
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(label_test, final_predictions)))
```
